```python
# randint is short for "random integer."
random.randint(-90,90)
```

```python
# randrange() function behaves differently than the previous two functions. 
# Inside the parentheses, we need to add two numbers, a lower and upper limit, separated by a comma.
# For the randrange() function, there is an option to add a step parameter and set it equal to an integer, which will generate increments of a given integer value, from the lower to the upper limit.
# random.randrange(-9,9,2)
random.randrange(-90,90, step=3)
```

```python
# uniform() function will allow us to generate a floating-point decimal number between two given numbers inside the parentheses.

random.uniform(-90,90)

```

```python
# Using these two functions, we can write an algorithm that will generate latitudes between -90 and 89. 
# Here is a small sample of what it might take to generate ten random floating-point decimal latitudes between -90 and 89.
x = 1
latitudes = []
while x < 11:
    random_lat = random.randint(-90, 89) + random.random()
    latitudes.append(random_lat)
    x += 1
latitudes
```

```python
# generate a floating-point decimal number between -90.000 and 90.000
# np.random.uniform(-90.000, 90.000)
np.random.uniform(low=-90, high=90)
```

```python
# add the size parameter when we use the NumPy module and set that equal to any whole number.

np.random.uniform(-90.000, 90.000, size=50)

```

```python
# To test how long a piece of code or function takes to run, 
# we can import the "timeit" module and use the %timeit magic command when we run our code or call the function.
%timeit np.random.uniform(-90.000, 90.000, size=1500)
# Using the np.random.uniform(-90.000, 90.000, size=1500) is 
# 100 times faster than using the function, and our code is one line, whereas the function uses eight lines!
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
# %timeit latitudes(1500)
```
